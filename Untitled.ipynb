```python
import numpy as np               # For numerical operations
import pandas as pd              # For data manipulation and analysis
import matplotlib.pyplot as plt  # For plotting graphs
import yfinance as yf            # For fetching stock price data from Yahoo Finance
```

```python

# Define the date range and stock ticker
start = '2012-01-01'
end = '2023-12-31'
stock = 'GOOG'

# Download stock data with auto adjustment for splits/dividends
data = yf.download(stock, start=start, end=end, auto_adjust=True)

# Check for missing values
print(data.isnull().sum())

# Handle missing values if necessary (e.g., forward fill)
data.fillna(method='ffill', inplace=True)

# Visualize the closing prices
plt.figure(figsize=(14, 5))
plt.plot(data['Close'], label='Adjusted Close Price')
plt.title(f'{stock} Stock Price History')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Reset index without adding old index as a column
data.reset_index(drop=True, inplace=True)

# Check for duplicates and remove them if necessary
data.drop_duplicates(inplace=True)

# Ensure proper data types (for example, converting 'Date' to datetime if needed)
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])

# Optionally print the first few rows to verify changes
print(data.head())
```

```python
# Assuming 'data' is your DataFrame containing stock prices
# Check available columns
print(data.columns)

# Ensure 'Date' is in datetime format if it exists
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])

# Visualize the closing prices and moving averages
plt.figure(figsize=(14, 7))

# Plot closing prices
if 'Date' in data.columns:
    plt.plot(data['Date'], data['Close'], label='Close Price', alpha=0.5)
else:
    # If 'Date' is not a column, use index
    plt.plot(data.index, data['Close'], label='Close Price', alpha=0.5)

# Plot moving average if it exists
if 'MA_100' in data.columns:
    plt.plot(data['MA_100'], label='100-Day MA', color='red')
else:
    print("MA_100 column not found in data.")

plt.title(f'{stock} Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()  # Optional: Add grid for better readability
plt.show()
```

```python
# Calculate the 100-day moving average and add it to the DataFrame
data['MA_100'] = data['Close'].rolling(window=100).mean()

# Calculate the 200-day moving average and add it to the DataFrame
data['MA_200'] = data['Close'].rolling(window=200).mean()

# Handle NaN values if necessary (e.g., forward fill)
data.fillna(method='bfill', inplace=True)  # Backward fill to handle NaNs

# Visualize the closing prices and moving averages
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', alpha=0.5)
plt.plot(data.index, data['MA_100'], label='100-Day MA', color='red')  # Keep 100-Day MA
plt.plot(data.index, data['MA_200'], label='200-Day MA', color='blue')    # Keep 200-Day MA

plt.title(f'{stock} Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Check the shape of the DataFrame before dropping NaNs
print(f"Shape before dropping NaNs: {data.shape}")

# Drop rows with NaN values
data.dropna(inplace=True)

# Check the shape after dropping NaNs
print(f"Shape after dropping NaNs: {data.shape}")

# Optionally, visualize missing data (if any)
import seaborn as sns

plt.figure(figsize=(14, 5))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Visualization')
plt.show()
```

```python
# Split the data into training and testing sets (80% train, 20% test)
train_size = int(len(data) * 0.80)

# Create training and testing DataFrames
data_train = data['Close'].iloc[:train_size].copy()
data_test = data['Close'].iloc[train_size:].copy()

# Optionally, print shapes to verify
print(f"Training data shape: {data_train.shape}")
print(f"Testing data shape: {data_test.shape}")
```

```python
# Get the number of samples in training and testing sets
num_train_samples = data_train.shape[0]
num_test_samples = data_test.shape[0]

# Print the number of samples
print(f"Number of training samples: {num_train_samples}")
print(f"Number of testing samples: {num_test_samples}")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
# Fit and transform the training data
data_train_scale = scaler.fit_transform(data_train.values.reshape(-1, 1))
# Optionally, check shapes of scaled data
print(f"Scaled training data shape: {data_train_scale.shape}")
```

```python
x = []
y = []

for i in range(100, data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
    
```

```python
x, y = np.array(x), np.array(y)
# Check the shapes of x and y
print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")
```

```python
from keras.layers import Dense, Dropout, LSTM, Input
from keras.models import Sequential
```

```python
# Initialize the model
model = Sequential()

# Add an Input layer followed by the first LSTM layer
model.add(Input(shape=(x.shape[1], 1)))  
model.add(LSTM(units=50, activation='tanh', return_sequences=True))
model.add(Dropout(0.2))

# Add subsequent LSTM layers without specifying input_shape
model.add(LSTM(units=100, activation='tanh', return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(units=150, activation='tanh', return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(units=200, activation='tanh'))
model.add(Dropout(0.5))

# Add the output layer
model.add(Dense(units=1)) 
```

```python
model.summary()
```

```python
# Compile the model with a lower learning rate
from keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)  # Adjust learning rate as necessary
model.compile(optimizer=optimizer, loss='mean_squared_error')
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.optimizers import Adam

# Initialize callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True)

# Compile the model with a lower learning rate
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Fit the model
history = model.fit(x, y, 
                    epochs=100,  # Adjust epochs as necessary
                    batch_size=32,  # Adjust batch size as necessary
                    validation_split=0.2,  # Use 20% of training data for validation
                    callbacks=[early_stopping, model_checkpoint])  # Include callbacks
```

```python
# Prepare x and y for predictions
x = []
y = []

for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i, 0])

x, y = np.array(x), np.array(y)
```

```python
# Make predictions
y_predict = model.predict(x)
```

```python
# Inverse transform to get actual prices
y_predict_inverse = scaler.inverse_transform(y_predict.reshape(-1, 1))

print(f"Predicted prices: {y_predict_inverse.flatten()}")
```

```python
# Plotting the results
plt.figure(figsize=(10, 8))
plt.plot(y_predict_inverse, 'r', label='Predicted Price')
plt.plot(y_inverse, 'g', label='Original Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Predicted vs Original Prices')
plt.legend()
plt.show()
```
